#Transformação e Qualificação dos dados

```python
#transforma em date timer e arredonda os milessegundos
df_customer['created'] = pd.to_datetime(df_customer['created'])
df_customer['created'] = df_customer['created'].dt.floor('S')
print(df_customer.dtypes)
```

```python
df_customer_null = df_customer.isnull().sum()
```

```python
df_customer_null
```

```python
#trocao o Null por nao informado
df_customer['genre'] = df_customer['genre'].fillna('nao informado')
```

```python
df_customer
```

```python
#transforma  para o formato de data
df_customer['birth']= pd.to_datetime(df_customer['birth'], format='%Y-%m-%d', errors='coerce')
print(df_customer.dtypes)
```

```python
now = datetime.now()
```

```python
#cria a coluna idade
df_customer['year'] = df_customer['birth'].apply(lambda x: now.year - x.year - ((now.month, now.day)< (x.month, x.day)))
```

```python
df_customer['birth'] = df_customer['birth'].fillna('nao informado')
```

```python
df_customer['year'] = df_customer['year'].fillna(0)
df_customer['year'] = df_customer['year'].astype(int)
```

```python
df_customer
```

```python
#Salva no formato csv
df_customer.to_csv('table_customers_trusted.csv', index=False)
```

#Tratamento dos dados da table_sales_items

```python
df_sales_items = pd.read_csv('table_sales_items.csv', encoding='utf-8')
```

```python
df_sales_items
```

```python
#Substitui os caracteres especiais
def substituir_acentos(text):
    substituicoes = {
        'Ã': 'A', 'â': 'a', 'Á': 'A', 'á': 'a',
        'À': 'A', 'à': 'a', 'Â': 'A', 'ã': 'a',
        'É': 'E', 'é': 'e', 'Ê': 'E', 'ê': 'e',
        'Í': 'I', 'í': 'i', 'Ó': 'O', 'ó': 'o',
        'Ô': 'O', 'ô': 'o', 'Õ': 'O', 'õ': 'o',
        'Ú': 'U', 'ú': 'u', 'Ç': 'C', 'ç': 'c'
    }

    for acentuado, substituto in substituicoes.items():
        text = text.replace(acentuado, substituto)
        text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df_sales_items['product_name'] = df_sales_items['product_name'].apply(substituir_acentos)
```

```python
df_sales_items
```

```python
#Salva o csv em partes pequenas para melhorar o ingestion no banco de dados
chunk_size = 1000

start_part = 11
end_part = 20

for i in range(0, len(df_sales_items), chunk_size):
  part = i // chunk_size + 1

  if part >= start_part and part <= end_part:
    df_chunk = df_sales_items.iloc[i:i+chunk_size]
    file_name = f'sales_items_part_{i // chunk_size + 1}.csv'

    df_chunk.to_csv(f'sales_items_part_{i // chunk_size + 1}.csv', index=False)
    print(f'Parte {i // chunk_size + 1} salva como sales_items_part_{i // chunk_size + 1}.csv')

    files.download(file_name)
  if part > end_part:
    break
```
